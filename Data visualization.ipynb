```python
import pandas as pd
```

```python
import numpy as np
```

```python
data=pd.read_csv("2019_nC0v_20200121_20200126_cleaned.csv")
```

```python
data
```

```python
import matplotlib.pyplot as pt
%matplotlib inline
```

# SCATTER PLOT

```python
corona1=pt.scatter(data["Confirmed"][:5],data["Suspected"][:5])
```

```python
pt.show(corona1)
```

# LINE PLOT

```python
corona2=pt.plot(data["Recovered"],data["Deaths"])
```

```python
pt.show(corona2)
```

# PIE PLOT

```python
corona3=pt.pie(data["Suspected"][:10],labels=data["Province/State"][:10])
```

```python
pt.show(corona3)
```

# BAR PLOT

```python
corona4=pt.bar(data["Province/State"],data["Suspected"])
```

```python
pt.show(corona4)
```
